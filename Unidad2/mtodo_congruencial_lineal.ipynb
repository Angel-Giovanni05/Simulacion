```python
import numpy as np
import matplotlib.pyplot as plt

def linear_congruential_generator(seed, a, c, m, n):
    """
    Genera una secuencia de números pseudoaleatorios usando el método del generador congruencial lineal.

    Parámetros:
    seed: int - Valor inicial (X0)
    a: int - Multiplicador
    c: int - Incremento
    m: int - Módulo
    n: int - Cantidad de números a generar

    Retorna:
    list - Secuencia de números generados
    """
    sequence = []
    Xn = seed
    for _ in range(n):
        Xn = (a * Xn + c) % m
        sequence.append(Xn)
    return sequence

# Parámetros iniciales
seed = 7    # Semilla inicial
a = 1664525  # Multiplicador
c = 1013904223  # Incremento
m = 2**32  # Módulo
n = 100  # Cantidad de números a generar

# Generar secuencia
random_numbers = linear_congruential_generator(seed, a, c, m, n)

# Normalizar valores entre 0 y 1 para graficar
normalized_numbers = [x / m for x in random_numbers]

# Mostrar la lista de números generados
print("Lista de números generados:")
for num in random_numbers:
    print(num)

# Visualización de la secuencia generada
plt.figure(figsize=(10, 4))
plt.plot(normalized_numbers, 'o-', markersize=4, label='Números pseudoaleatorios')
plt.xlabel('Índice')
plt.ylabel('Valor (normalizado)')
plt.title('Generador Congruencial Lineal')
plt.legend()
plt.grid()
plt.show()
```
