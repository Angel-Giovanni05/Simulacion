##Problema 2: Distribución Exponencial
#Contexto: Una estación de carga rápida para vehículos eléctricos quiere
simular el tiempo entre la llegada de los vehículos para prever la cantidad
de puntos de carga necesarios durante el día. La llegada de los vehículos
sigue una distribución exponencial con una tasa de 1 llegada cada 15
minutos, en promedio.
Objetivo: Generar tiempos entre llegadas de vehículos eléctricos durante
un día de operación (12 horas).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución exponencial
llegadas_por_minuto = 1 / 15  # Tasa de llegadas (1 cada 15 minutos)
tiempo_operacion = 12 * 60  # 12 horas en minutos

# Simulación de tiempos entre llegadas
intervalos_llegadas = np.random.exponential(1 / llegadas_por_minuto, tiempo_operacion)

# Calcular tiempos absolutos de llegada
tiempos_llegada = np.cumsum(intervalos_llegadas)

# Filtrar tiempos dentro del período de operación (12 horas)
tiempo_maximo = 12 * 60
tiempo_llegada_filtrado = tiempos_llegada[tiempos_llegada <= tiempo_maximo]

# Mostrar los primeros tiempos de llegada
print("Tiempos de llegada de los primeros vehículos en minutos:", tiempo_llegada_filtrado[:20])

# Visualización de los datos
plt.figure(figsize=(8,5))
plt.hist(intervalos_llegadas, bins=30, color='lightblue', edgecolor='black', density=True)
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Densidad")
plt.title("Distribución de tiempos entre llegadas de vehículos")
plt.show()

# Gráfico de los tiempos de llegada
plt.figure(figsize=(8,5))
plt.plot(tiempo_llegada_filtrado, np.arange(1, len(tiempo_llegada_filtrado)+1), marker='o', linestyle='-', color='green')
plt.xlabel("Tiempo (minutos)")
plt.ylabel("Número de llegadas acumuladas")
plt.title("Llegadas acumuladas de vehículos eléctricos")
plt.show()
```
